```python
import subprocess

def read_nfcapd_file(file_path):
    try:
        result = subprocess.run(['nfdump', '-r', file_path], capture_output=True, text=True)
        if result.returncode == 0:
            print(result.stdout)
        else:
            print(f"Error reading file: {result.stderr}")
    except Exception as e:
        print(f"An error occurred: {e}")

#  
read_nfcapd_file('nfcapd.202409252000')
```
